```python
import nltk
from nltk.chat.util import Chat, reflections
import random


pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, how can I help you today?",]
    ],
    [
        r"hi|hello|hey",
        ["Hello!", "Hi there!", "Hey! How can I assist you?",]
    ],
    [
        r"what is your name ?",
        ["I am a chatbot created to assist you with your queries.",]
    ],
    [
        r"how are you ?",
        ["I'm just a program, but thanks for asking! How can I help you?",]
    ],
    [
        r"(.*) help (.*)",
        ["I can help you with programming, information, or just have a chat! What do you need?",]
    ],
    [
        r"(.*) created you?",
        ["I was created by a programmer as part of a Python task.",]
    ],
    [
        r"tell me a joke",
       ["Why did the math book look sad?Because it had too many problems!", 
         "Why did the scarecrow win an award? Because he was outstanding in his field!"]
        
    ],
    [
        r"what can you do?",
        ["I can chat with you, answer basic questions, and tell jokes! What would you like to know?",]
    ],
    [
        r"quit",
        ["Goodbye! Have a nice day!", "It was nice talking to you. See you later!"]
    ],
]


chatbot = Chat(pairs, reflections)


print("Hi! I'm your chatbot. Type 'quit' to end the conversation.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        print("Chatbot: Goodbye! Have a nice day!")
        break
    response = chatbot.respond(user_input)
    if response:
        print("Chatbot:", response)
    else:
        
        fallback_responses = [
            "I'm sorry, I don't understand that. Can you rephrase?",
            "That's interesting! Can you tell me more?",
            "I'm not sure I follow. Could you clarify?",
            "Let's talk about something else! What interests you?"
        ]
        print("Chatbot:", random.choice(fallback_responses)) 
```
